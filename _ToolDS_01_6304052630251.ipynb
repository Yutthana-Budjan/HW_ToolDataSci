## Dataset

```python
import pandas as pd
header = ["Number_of_times_pregnant", #Feature_1
          "Plasma_glucose_concentration_a_2-hours_in_an_oral_glucose_tolerance_test", #Feature_2
          "Diastolic_blood_pressure", #Feature_3
          "Triceps_skin_fold_thickness", #Feature_4
          "2-Hour_serum_insulin", #Feature_5
          "Body_mass_index", #Feature_6
          "Diabetes_pedigree_function", #Feature_7
          "Age", #Feature_8
          "Class_variable"] #Labels or Class
df = pd.read_csv("pima-indians-diabetes.csv", names=header)
df
```

## RFE (Logistic Regression)

```python
import pandas as pd
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
X = df.drop("Class_variable",axis=1)
y = df["Class_variable"]
model = LogisticRegression()
rfe = RFE(model, n_features_to_select=4)
fit = rfe.fit(X, y)

# Create DataFrame for Summary
print("Number Features: %s" % (fit.n_features_))
feature_names = [f"Feature_{i+1}" for i in range(X.shape[1])]
df_results = pd.DataFrame({
    "Feature_Name": feature_names,
    "Feature_Ranking": fit.ranking_,
    "Select_Feature": fit.support_})
print(df_results)
```

```python
# Test Model RFE
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
selected_column_rfe = ["Number_of_times_pregnant", #Feature_1
                      "Plasma_glucose_concentration_a_2-hours_in_an_oral_glucose_tolerance_test", #Feature_2
                      "Body_mass_index", #Feature_6
                      "Diabetes_pedigree_function", #Feature_7
                      "Class_variable"] #Labels or Class]
df_rfe = df[selected_column_rfe]

X_train, X_test, y_train, y_test = train_test_split(df_rfe.drop("Class_variable", axis=1), df_rfe["Class_variable"], test_size=0.2, random_state=40)
model_rfe = LogisticRegression()
model_rfe.fit(X_train, y_train)
y_pred = model_rfe.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')
```

## Statistic Feature Selection (ANOVA)

```python
import pandas as pd
from sklearn.datasets import make_classification
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif
from matplotlib import pyplot
X = df.drop("Class_variable",axis=1)
y = df["Class_variable"]
feature_selection = SelectKBest(score_func=f_classif, k=4)
X_selected = feature_selection.fit_transform(X, y)

# Create DataFrame for Summary
print("Original Number of Features:", X.shape[1])
print("Selected Features:", X_selected.shape[1])
feature_names = [f"Feature_{i+1}" for i in range(X.shape[1])]
df_results = pd.DataFrame({
    "Feature_name": feature_names,
    "F-statistic": feature_selection.scores_})
df_results["Feature_Ranking"] = df_results["F-statistic"].rank(ascending=False).astype(int)
df_results["Select_Feature"] = df_results["Feature_Ranking"] <= X_selected.shape[1]
print(df_results)

pyplot.bar([x for x in range(len(feature_selection.scores_))], feature_selection.scores_)
pyplot.show()
```

```python
# Test Model Statistics Feature Selection
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
selected_column_stat = ["Number_of_times_pregnant", #Feature_1
                        "Plasma_glucose_concentration_a_2-hours_in_an_oral_glucose_tolerance_test", #Feature_2
                        "Body_mass_index", #Feature_6
                        "Age", #Feature_8
                        "Class_variable"] #Labels or Class]
df_stat = df[selected_column_stat]

X_train, X_test, y_train, y_test = train_test_split(df_stat.drop("Class_variable", axis=1), df_stat["Class_variable"], test_size=0.2, random_state=40)
model_stat = LogisticRegression()
model_stat.fit(X_train, y_train)
y_pred = model_stat.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')
```

## Feature Importance (Decision Tree)

```python
import pandas as pd
from sklearn.datasets import make_regression
from sklearn.tree import DecisionTreeRegressor
from matplotlib import pyplot
X = df.drop("Class_variable",axis=1)
y = df["Class_variable"]
model = DecisionTreeRegressor()
model.fit(X, y)
feature_importance = model.feature_importances_

# Create DataFrame for Summary
feature_names = [f"Feature_{i+1}" for i in range(X.shape[1])]
df_results = pd.DataFrame({
    "Feature_name": feature_names,
    "Score": feature_importance})
df_results["Feature_Ranking"] = df_results["Score"].rank(ascending=False).astype(int)
df_results["Select_Feature"] = df_results["Feature_Ranking"] <= 4
print(df_results)

pyplot.bar([x for x in range(len(feature_importance))], feature_importance)
pyplot.show()
```

```python
# Test Model Feature Importance
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
selected_column_imp = [ "Plasma_glucose_concentration_a_2-hours_in_an_oral_glucose_tolerance_test", #Feature_2
                        "Body_mass_index", #Feature_6
                        "Diabetes_pedigree_function", #Feature_7
                        "Age", #Feature_8
                        "Class_variable"] #Labels or Class]
df_imp = df[selected_column_imp]

X_train, X_test, y_train, y_test = train_test_split(df_imp.drop("Class_variable", axis=1), df_imp["Class_variable"], test_size=0.2, random_state=40)
model_imp = LogisticRegression()
model_imp.fit(X_train, y_train)
y_pred = model_imp.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')
```
